5. A gourmet cafe serves exotic salmon at lunch and dinner with the number of customers ordering them as follows:

Lunch:          0   1   2

Probability     0.3 0.5 0.2


Dinner:          0   1   2

Probability     0.2 0.4 0.4

What is the expected total demand for the fish in a day?

**SOLUTION**

Expected total demand = Total demand at lunch + Total demand at Dinner

total demand at Lunch = 0(0.3) + 1(0.5) + 2(0.2) = 0.9
total demand at Dinnner = 0(0.2) + 1(0.4) + 2(0.4) = 1.2
therefore:

Expected Total Demand = 0.9 + 1.2 = 2.10

E**xpected Total Demand = 2.10**
